# CSV to JSON

```python
import pandas as pd

df = pd.read_csv("norming-targets.csv")

df = df[["name", "morality", "intro", "description", "motive"]]
df.to_json("norming-targets.json", orient="records")
```

# Print Table

```python
import pandas as pd

df = pd.read_csv("norming-targets.csv").sort_values(by=['name']).sort_values(by=['morality'], ascending=False)
df['stimulus'] = df['stimulus'].str.strip('\n')


df[['name', 'morality', 'stimulus']].to_csv('../../../curiosity-explain/study2/data/study2-targets.csv', index=False)
```
